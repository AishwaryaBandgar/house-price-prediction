```python
import numpy as np 
import pandas as pd 
```

```python
np.random.seed(42)
data_size = 10000

Area = np.random.randint(500, 4000, data_size) # values between these(500, 4000), data_size(no. of values topic)
Bedrooms = np.random.randint(1, 6, data_size)  
House_Age = np.random.randint(0, 30, data_size) # random int
Distance = np.random.uniform(1, 20, data_size) # random floatvalues 
```

```python
# price depends on feature (Real life)

Price = ( Area * 0.03 + 
         Bedrooms * 5 - 
         House_Age * 0.7 - 
         Distance * 1.5 +
         np.random.normal(0, 10, data_size) ) # noise

df = pd.DataFrame({ "Area" : Area,
                    "Bedrooms" : Bedrooms,
                    "House_Age" : House_Age,
                    "Distance" : Distance,
                    "Price" : Price })
                   
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
X = df.drop("Price", axis=1)
y = df["Price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state =  42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MSE :", mse)
print("RMSE:", rmse)
print("R2  :", r2)

```

```python
def predict_price(area, bedrooms, age, distance):
    user_df = pd.DataFrame(
        [[area, bedrooms, age, distance]],
        columns=["Area", "Bedrooms", "House_Age", "Distance"]
    )
    return model.predict(user_df)[0]

area = float(input("Enter Area (sq ft): "))
bedrooms = int(input("Enter Number of Bedrooms: "))
house_age = int(input("Enter House Age (years): "))
distance = float(input("Enter Distance from city (km): "))

print("Predicted Price (Lakhs):", round(predict_price(area, bedrooms, house_age, distance), 2))
```
